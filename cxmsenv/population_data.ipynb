```python
import pandas as pd

csv_file_path = '../../world_population_data.csv'
world_population_df = pd.read_csv(csv_file_path)

world_population_df.head()

def construct_image_path(country_name):
    normalized_name = country_name.replace(" ", "_").replace(",", "").replace("'", "")
    image_path = f"images/games-images/country-outlines/{normalized_name}.png"
    return image_path

world_population_df['image_path'] = world_population_df['country'].apply(construct_image_path)

world_population_df.head()
```

```python
world_population_df.shape
```

```python
import os

def image_exists(country_name):
    normalized_name = country_name.replace(" ", "_").replace(",", "").replace("'", "")
    image_path = f"../static/images/games-images/country-outlines/{normalized_name}.png"
    return os.path.exists(image_path), f"images/games-images/country-outlines/{normalized_name}.png"

# Apply the check and create image paths only for existing images
world_population_df['image_exists'], world_population_df['image_path'] = zip(*world_population_df['country'].apply(image_exists))

# Filter out countries without an image
filtered_df = world_population_df[world_population_df['image_exists']].drop('image_exists', axis=1)
```
